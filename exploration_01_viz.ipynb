```python
input_shape = (64,64,1)
stacked_encoder = keras.models.Sequential([
    keras.Input(shape=input_shape),
    keras.layers.Flatten(),
    keras.layers.Dense(2000,activation = 'selu'),
    keras.layers.Dense(1000,activation = 'selu'),
    keras.layers.Dense(600,activation = 'selu'),
    keras.layers.Dense(400,activation = 'selu'),
    keras.layers.Dense(100,activation = 'selu'),
    keras.layers.Dense(30,activation = 'selu'),
])
# tf.keras.layers.InputLayer(input_shape=(4,))
stacked_decoder = keras.models.Sequential([
    keras.layers.Dense(100,activation = 'selu',input_shape=(30,)),
    keras.layers.Dense(400,activation = 'selu'),
    keras.layers.Dense(600,activation = 'selu'),
    keras.layers.Dense(1000,activation = 'selu'),
    keras.layers.Dense(2000,activation = 'selu'),
    keras.layers.Dense(64*64,activation = 'linear'),
    keras.layers.Reshape((64,64)),
])

autoencoder = keras.models.Sequential([stacked_encoder,stacked_decoder])
autoencoder.compile(loss="mean_absolute_error",optimizer = "adam")
```

```python
def show_reconstructions(model, nb_sample=5):
    
    reconstructions = model.predict(X_val[:nb_sample])
    fig = plt.figure(figsize=(12,12))
    for idx in range(nb_sample):
        plt.subplot(2,nb_sample,1+idx)
        plt.imshow(X_val[idx],cmap="gray")
        plt.subplot(2,nb_sample,1+nb_sample+idx)
        plt.imshow(reconstructions[idx],cmap="gray")
```
